```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
DATA_PATH = "telecom_churn.csv"   # change if needed
df = pd.read_csv(DATA_PATH)
```

```python
print("Initial shape:", df.shape)
display(df.head())
```

```python
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
```

```python
target_candidates = [c for c in df.columns if "churn" in c]
if not target_candidates:
    raise KeyError("No churn column found. Make sure your dataset has 'Churn'.")
target = target_candidates[0]
print("Target column detected:", target)
if df[target].dtype == "object":
    df[target] = df[target].str.strip().str.lower().map({
        "yes": 1, "no": 0,
        "true": 1, "false": 0,
        "1": 1, "0": 0
    })
df[target] = df[target].astype(int)

print("\nChurn distribution:")
print(df[target].value_counts(normalize=True))
```

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
if target in num_cols:
    num_cols.remove(target)

cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()

print("\nNumeric columns:", num_cols)
print("Categorical columns:", cat_cols)

num_imputer = SimpleImputer(strategy="median")
cat_imputer = SimpleImputer(strategy="most_frequent")

df[num_cols] = num_imputer.fit_transform(df[num_cols])
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

print("\nMissing values after cleaning:", df.isnull().sum().sum())
```

```python
if "tenure" in df.columns:
    df["tenure_bin"] = pd.cut(df["tenure"], bins=[-1, 11, 35, 1000],
                              labels=["short", "medium", "long"])

if "monthlycharges" in df.columns:
    df["monthly_bin"] = pd.qcut(df["monthlycharges"], q=3,
                                labels=["low", "medium", "high"])

# === Step 6: Save cleaned dataset ===
CLEAN_PATH = "telecom_churn_cleaned.csv"
df.to_csv(CLEAN_PATH, index=False)
print(f"\n✅ Cleaned dataset saved to: {CLEAN_PATH}")
print("Final shape:", df.shape)

df.head()
```
